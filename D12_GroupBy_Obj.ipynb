# GroupBy Object

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('Datasets/datasets-session-19/imdb-top-1000.csv')
movies.head()
```

```python
genres = movies.groupby('Genre')
```

```python
# Applying builtin aggregation fuctions on groupby objects
genres.sum()
```

```python
# find the top 3 genres by total earning
movies.groupby('Genre')
movies.groupby('Genre').sum()
movies.groupby('Genre').sum()['Gross']
movies.groupby('Genre').sum()['Gross'].sort_values()
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
# same as above, just different ordering
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

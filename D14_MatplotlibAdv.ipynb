```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('Datasets/datasets-session-24/iris.csv')
iris.sample(5)
```

```python
# Labelling Species with numbers

iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species']) # different colors to see different flowers
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Plot size

```python
plt.figure(figsize=(15,7)) # (width, height)

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Annotations

```python
batter = pd.read_csv('Datasets/datasets-session-24/batter.csv')
batter
```

```python
sample_df = batter.head(100).sample(25, random_state=5)
sample_df
```

```python
# small example
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3',fontdict={'size':15,'color':'red'})
plt.text(4,8,'Point 4',fontdict={'size':12,'color':'brown'})
```

```python
plt.figure(figsize=(17, 8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i]) # just like above small example -> [plt.text(1,5,'Point 1')]
```

### Horizontal and Vertical lines

```python
# Eg. choosing the batsmen whose strike rate is greater than 130

plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')    # for strike rate greater than 130
plt.axvline(30,color='red')     # for average greater than 140
plt.axhline(140,color='green')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,6))                    # there are 2 axes in this code

ax[0,0].scatter(batter['avg'], batter['strike_rate'], color='red')    # 1st graph
ax[0,1].scatter(batter['avg'], batter['runs'])                        # 2nd graph
ax[1,0].hist(batter['avg'])                                           # 3rd graph
ax[1,1].hist(batter['avg'])                                           # 4th graph
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### 3D Contour Plots

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

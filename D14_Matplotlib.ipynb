```python
# import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

### 2D Line plot

```python
'''
1. Bivariate Analysis      (2D plot is always plotted on 2 columns among multiple columns)

2. [categorical, numerical] and [numerical, numerical]
    1 column      another        1 col        another

3. Use case - Time series data       (1 quantity is measured on Time axis) eg. stocks or revenue
'''
```

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]  # numerical column
year = [2015,2016,2017,2018,2019,2020]         # categorical column

plt.plot(year, price)  # (x-axis, y-axis) 
```

```python
batsman = pd.read_csv('Datasets/datasets-session-23/sharma-kohli.csv')
batsman
# data of sharma and kohli with total runs in every season
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='upper right')
```

```python
# limiting axes - used when there's a outlier
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)   # trim price 0 to 75000
plt.xlim(2016,2020) # trim year 2016 to 2020
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
# plt.show() - used to plot the graph on some different environments
```

### Scatter Plots

```python
''' 
1. Bivariate Analysis   -  required two columns
2. numerical vs numerical  -  mostly used on numerical vs numerical data
3. Use case - Finding correlation
'''
```

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)
x
y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x, y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('Datasets/datasets-session-23/batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'], df['strike_rate'], color='red', marker='+')
plt.title('Avg and SR analysis of Top Batsman')
plt.xlabel('Average')
plt.ylabel('Strike rate')
```

```python
# marker
```

```python
# size
tips = sns.load_dataset('tips')
tips.head()
```

```python

# slower
plt.scatter(tips['total_bill'], tips['tip'] ,s=tips['size']*20) # 's' makes changes in dots according to the size values
# more people = bigger dots
```

```python
# scatterplot using plt.plot
# faster in compared to plt.scatter - because cannot use attributes like color, marker, etc
plt.plot(tips['total_bill'],tips['tip'],'o')
```

### Bar chart

```python
'''
1. Bivariate Analysis
2. Numerical vs Categorical
3. Use case - Aggregate analysis of groups
'''
```

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors, children, color='orange')
```

```python
plt.barh(colors, children, color='red')
```

```python
df = pd.read_csv('Datasets/datasets-session-23/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'])
```

```python
plt.bar(df['batsman'], df['2015'], width=0.5)
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show() # also removes unwanted data
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']
plt.bar(colors,children,color='black')
```

```python
plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')
plt.legend()
plt.show()
```

### Pie chart

```python
''' 
Univariate/Bivariate Analysis
Categorical vs numerical
Use case - To find contibution on a standard scale
'''
```

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# dataset 
df = pd.read_csv('Datasets/datasets-session-23/gayle-175.csv') # match of rcb in which Chris Gayle scored 175
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', explode=[0.3,0,0,0,0,0.1])
plt.show()
```
